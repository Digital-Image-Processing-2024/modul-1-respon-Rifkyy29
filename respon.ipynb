```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2 as cv
```

### 1. RGB Color Image

```python
image = img.imread("respon.jpg")
plt.imshow(image)
```

```python
rgb_image = cv.cvtColor(image,cv.COLOR_BGR2RGB)

red_channel = rgb_image[:, :, 0]
green_channel = rgb_image[:, :, 0]
blue_channel = rgb_image[:, :, 0]

plt.imshow(rgb_image)
```

### Red Colour Image

```python
red_image = np.copy(rgb_image)
red_image[:,:,1] = 0
red_image[:,:,2] = 0

plt.imshow(red_image)
```

### Red Histogram

```python
x, y = red_channel.shape

sum = np.zeros(256).astype(int)
for i in range(x):
    for j in range(y):
        val = red_channel[i,j]
        sum[val] +=1 

# inisialisasi array
red_imgValue = [i for i in range (0,256)]
plt.bar(red_imgValue, sum/(x*y), color = "red")
plt.xlabel('Intensitas Piksel')
plt.ylabel('Frekuensi')
plt.show()
```

### 2.Zoom In Zoom Out Image

```python
grayimage = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(grayimage,cmap="gray")
```

```python
def zoom_in(image, zoom_factor):
    height, width = image.shape[:2]
    new_height = int(height * zoom_factor)
    new_width = int(width * zoom_factor)
    zoomed_image = np.zeros((new_height, new_width, 3), dtype=np.uint8)
    for i in range(new_height):
        for j in range(new_width):
            zoomed_image[i, j] = image[int(i / zoom_factor), int(j / zoom_factor)]
    return zoomed_image

def zoom_out(image, zoom_factor):
    height, width = image.shape[:2]
    new_height = int(height / zoom_factor)
    new_width = int(width / zoom_factor)
    zoomed_image = np.zeros((new_height, new_width, 3), dtype=np.uint8)
    for i in range(new_height):
        for j in range(new_width):
            zoomed_image[i, j] = image[int(i * zoom_factor), int(j * zoom_factor)]
    return zoomed_image

zoom_in_image = zoom_in(grayimage, 2)
zoom_out_image = zoom_out(grayimage, 2)

```

```python
plt.imshow(zoom_in_image)
plt.title('Zoom In')
plt.show()
```

```python
plt.imshow(zoom_out_image)
plt.title('Zoom Out')
plt.show()
```

Untuk melihat perbedaan nya lebih jelas dapat menggunakan CV

```python
cv.imshow('Zoom In', zoom_in_image)
cv.imshow('Zoom Out', zoom_out_image)
cv.waitKey(0)
```
